```python
import numpy as np 
import os
import cv2
from google.colab.patches import cv2_imshow
import cv2 as cv
from matplotlib import pyplot as plt
import skimage
from scipy import ndimage as ndi
%matplotlib inline
from google.colab import files
```

Read the given images

```python
frame1 = cv2.imread("./3.jfif", 0)
cv2_imshow(frame1)
frame2 = cv2.imread("./8.png")
cv2_imshow(frame2)
```

#PREWITT FILTER

```python
prewitt_kernel_on_x = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
prewitt_kernel_on_y = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
```

#ROBERTS FILTER

```python
roberts_kernel_on_x = np.array([[-1,0],[0,1]])
roberts_kernel_on_y = np.array([[0,-1],[1,0]])
```

#LAPLACIAN FILTER

```python
result_laplacian_filter_image1 = cv2.Laplacian(frame1,cv2.CV_64F, borderType=cv2.BORDER_DEFAULT)
result_laplacian_filter_image2 = cv2.Laplacian(frame2,cv2.CV_64F, borderType=cv2.BORDER_DEFAULT)
```

#SOBEL FILTER

```python
result_sobel_filter_on_x_image1   = cv2.Sobel(frame1,cv2.CV_64F,1,0,ksize=5,borderType=cv2.BORDER_DEFAULT )
result_sobel_filter_on_y_image1   = cv2.Sobel(frame1,cv2.CV_64F,0,1,ksize=5, borderType=cv2.BORDER_DEFAULT)
result_sobel_magnitude_image1     = np.sqrt(result_sobel_filter_on_x_image1**2 + result_sobel_filter_on_y_image1**2)

result_sobel_filter_on_x_image2   = cv2.Sobel(frame2,cv2.CV_64F,1,0,ksize=5, borderType=cv2.BORDER_DEFAULT)
result_sobel_filter_on_y_image2   = cv2.Sobel(frame2,cv2.CV_64F,0,1,ksize=5, borderType=cv2.BORDER_DEFAULT)
result_sobel_magnitude_image2     = np.sqrt(result_sobel_filter_on_x_image2**2 + result_sobel_filter_on_y_image2**2)
```

#COMPUTE AND ANALYZE FILTERS

```python
result_prewitt_filter_on_x_image1 = cv2.filter2D(frame1, -1, prewitt_kernel_on_x)
result_prewitt_filter_on_y_image1 = cv2.filter2D(frame1, -1, prewitt_kernel_on_y)

result_prewitt_filter_on_x_image2 = cv2.filter2D(frame2, -1, prewitt_kernel_on_x)
result_prewitt_filter_on_y_image2 = cv2.filter2D(frame2, -1, prewitt_kernel_on_y)

result_roberts_filter_on_x_image1 = cv2.filter2D(frame1, -1, roberts_kernel_on_x)
result_roberts_filter_on_y_image1 = cv2.filter2D(frame1, -1, roberts_kernel_on_y)

result_roberts_filter_on_x_image2 = cv2.filter2D(frame2, -1, roberts_kernel_on_x)
result_roberts_filter_on_y_image2 = cv2.filter2D(frame2, -1, roberts_kernel_on_y)
```

```python
fig, axes = plt.subplots(nrows=9, ncols=2, figsize=(frame1.shape[1], frame1.shape[0]), sharex=True, sharey=True)

axes[0][0].imshow(frame1,cmap = 'gray')
axes[0][0].set_title('Original Img 1',fontsize=100)
axes[0][1].imshow(frame2,cmap = 'gray')
axes[0][1].set_title('Original Img 2',fontsize=100)

axes[1][0].imshow(result_prewitt_filter_on_x_image1,cmap = 'gray')
axes[1][0].set_title('Prewitt on X Img 1',fontsize=100)
axes[1][1].imshow(result_prewitt_filter_on_x_image2,cmap = 'gray')
axes[1][1].set_title('Prewitt on X Img 2',fontsize=100)

axes[2][0].imshow(result_prewitt_filter_on_y_image1,cmap = 'gray')
axes[2][0].set_title('Prewitt on Y Img 1',fontsize=100)
axes[2][1].imshow(result_prewitt_filter_on_y_image2,cmap = 'gray')
axes[2][1].set_title('Prewitt on Y Img 2',fontsize=100)

axes[3][0].imshow(result_roberts_filter_on_x_image1,cmap = 'gray')
axes[3][0].set_title('Roberts on X Img 1',fontsize=100)
axes[3][1].imshow(result_roberts_filter_on_x_image2,cmap = 'gray')
axes[3][1].set_title('Roberts on X Img 2',fontsize=100)

axes[4][0].imshow(result_roberts_filter_on_y_image1,cmap = 'gray')
axes[4][0].set_title('Roberts on Y Img 1',fontsize=100)
axes[4][1].imshow(result_roberts_filter_on_y_image2,cmap = 'gray')
axes[4][1].set_title('Roberts on Y Img 2',fontsize=100)

axes[5][0].imshow(result_sobel_filter_on_x_image1,cmap = 'gray')
axes[5][0].set_title('Sobel on X Img 1',fontsize=100)
axes[5][1].imshow(result_sobel_filter_on_x_image2,cmap = 'gray')
axes[5][1].set_title('Sobel on X Img 2',fontsize=100)

axes[6][0].imshow(result_sobel_filter_on_y_image1,cmap = 'gray')
axes[6][0].set_title('Sobel on Y Img 1',fontsize=100)
axes[6][1].imshow(result_sobel_filter_on_y_image2,cmap = 'gray')
axes[6][1].set_title('Sobel on Y Img 2',fontsize=100)

axes[7][0].imshow(result_sobel_magnitude_image1,cmap = 'gray')
axes[7][0].set_title('Sobel Magnitude Img 1',fontsize=100)
axes[7][1].imshow(result_sobel_magnitude_image2,cmap = 'gray')
axes[7][1].set_title('Sobel Magnitude Img 2',fontsize=100)

axes[8][0].imshow(result_laplacian_filter_image1,cmap = 'gray')
axes[8][0].set_title('Laplacian Img 1',fontsize=100)
axes[8][1].imshow(result_laplacian_filter_image2,cmap = 'gray')
axes[8][1].set_title('Laplacian Img 2',fontsize=100)
```
